```python
df2 = df.drop(['hhid', 'clust', 'pid', 's10aq4', 's10aq6', 's10aq7', 's10aq8', 's10aq9', 'strata_id', 'weight', 'pov_jan', 'epov_jan'], axis='columns')
df2.head()
```

```python
df.poverty.unique()
```

```python
# Assign numerical value to each category

df2['poverty_converted'] = df2.poverty.str.strip().str.lower().map({
    "non poor": 2,
    "moderately poor": 1,
    "severaly poor": 0,
})
```

```python
df2['s10aq1_converted'] = df2.s10aq1.apply(lambda x: 1 if x == 'Yes' else 0)
df2.head()
```

```python
# compute class imbalance

counts = df2['s10aq1_converted'].value_counts()
print(counts)
```

```python
majority = counts.max()
minority = counts.min()

imbalance = majority / minority
print(f"Imbalance ration {imbalance}") # how cone class appear much more than another
```

```python
df3 = df2.drop('s10aq2', axis='columns')
df3.head()
```

```python
# Data cleaning cont..
df3['s10aq3_converted'] = df3.s10aq3.apply(lambda x: 1 if x == 'Yes' else 0)
df3.head(3)
```

```python
# On hot encoder for province and district

dummies = pd.get_dummies(df3.province).astype(int)
dummies.head()
```

```python
df4 = pd.concat([df3, dummies], axis='columns')
df4.head()
```

```python
district_dummies = pd.get_dummies(df4.district).astype(int)
district_dummies.head()

df5 = pd.concat([df4, district_dummies], axis='columns')
df5.head(3)
```

```python
df5['ur_converted'] = df5.ur.apply(lambda x: 1 if x == 'Urban' else 0)
df5.head(3)
```

```python
df5.quintile.unique().sum()
```

```python
# Assign numerical value to each category

df5['quintile_converted'] = df5.quintile.str.strip().str.lower().map({
    "poor": 0,
    "poor/mid": 0.5,
    "mid": 1,
    "mid/upper": 1.5,
    "upper": 2
})
```
